## Изучение данных из файла

```python
#чтение датафрейма, вывод части таблицы и инфо
import pandas as pd
data = pd.read_csv('/datasets/real_estate_data.csv', sep='\t')
display(data.head(20))
data.info()
```

### Вывод

В столбцах ceiling_height, floors_total, living_area, is_apartment, kitchen_area, balcony, locality_name, airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition имеются пропуски

## Предобработка данных

```python
#Оценка количества пропусков и уникальных значений в столбце floors_total
floors_total_proportion = data['floors_total'].isna().sum() / data['floors_total'].count()
print(f'Доля пропусков в столбце floors_total: {floors_total_proportion:.1%}')
data['floors_total'].unique()
```

```python
#Замена пропусков медианным значением и типа переменной на int
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['floors_total'] = data['floors_total'].astype('int')
```

```python
#Оценка количества пропусков и уникальных значений в столбце ceiling_height
ceiling_height_proportion = data['ceiling_height'].isna().sum() / data['ceiling_height'].count()
print(f'Доля пропусков в столбце ceiling_height: {ceiling_height_proportion:.0%}')
data['ceiling_height'].unique()
```

```python
#Замена значений-выбросов, после чего заполнение пропусков средним значением
data.loc[data['ceiling_height'] > 9, 'ceiling_height'] = data['ceiling_height'] / 10
data.loc[(data['ceiling_height'] < 2) | (data['ceiling_height'] > 5), 'ceiling_height'] = data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].mean())
```

```python
#Оценка количества пропусков и уникальных значений в столбце balcony
balcony_proportion = data['balcony'].isna().sum() / data['balcony'].count()
print(f'Доля пропусков в столбце balcony: {balcony_proportion:.0%}')
data['balcony'].unique()
```

```python
#Замена пропусков на 0 и изменение типа переменной на int
data['balcony'] = data['balcony'].fillna(0)
data['balcony'] = data['balcony'].astype('int')
```

```python
#Замена пропусков на значение False, изменение типа переменной на bool
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'] = data['is_apartment'].astype('bool')
```

```python
#Оценка количества пропусков и уникальных значений в столбце kitchen_area
kitchen_area_proportion = data['kitchen_area'].isna().sum() / data['kitchen_area'].count()
print(f'Доля пропусков в столбце kitchen_area: {kitchen_area_proportion:.0%}')
data['kitchen_area'].unique()
```

```python
#Расчет доли кухни в общей площади и замена пропусков на медианное умноженное на общую площадь
kitchen_proportion = data['kitchen_area'] / data['total_area']
data['kitchen_area'] = data['kitchen_area'].fillna(kitchen_proportion.median() * data['total_area'])
```

```python
#Оценка количества пропусков и уникальных значений в столбце living_area
living_area_proportion = data['living_area'].isna().sum() / data['living_area'].count()
print(f'Доля пропусков в столбце living_area: {living_area_proportion:.0%}')
data['living_area'].unique()
```

```python
#Замена пропусков на разность между общей площадью и площадью кухни
data['living_area'] = data['living_area'].fillna(data['total_area'] - data['kitchen_area'])
```

```python
#Оценка количества пропусков и уникальных значений в столбце parks_around3000
parks_around3000_proportion = data['parks_around3000'].isna().sum() / data['parks_around3000'].count()
print(f'Доля пропусков в столбце parks_around3000: {parks_around3000_proportion:.0%}')
data['parks_around3000'].unique()
```

```python
#Предположим, что все отсутствующие значения равны 0 и заменим тип переменной на int
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['parks_around3000'] = data['parks_around3000'].astype('int')
```

```python
#Оценка количества пропусков и уникальных значений в столбце ponds_around3000
ponds_around3000_proportion = data['ponds_around3000'].isna().sum() / data['ponds_around3000'].count()
print(f'Доля пропусков в столбце ponds_around3000: {ponds_around3000_proportion:.0%}')
data['ponds_around3000'].unique()
```

```python
#Предположим, что все отсутствующие значения равны 0 и заменим тип переменной на int
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].astype('int')
```

```python
#Предположим, что объявление еще не снято и заменим пропуски количество дней между сегодняшней датой и датой размещения
from datetime import *
data['first_day_exposition'] = data['first_day_exposition'].map(lambda x: datetime.strptime(x, "%Y-%m-%dT%H:%M:%S").date())
now = date.today()
#print(now, data['first_day_exposition'])
data['days_exposition'] = data['days_exposition'].fillna((now - data['first_day_exposition']).astype("str").map(lambda x: x.split()).str[0].astype(int))
```

```python
#Оценка доли пропусков названии населенных пунктов
locality_name_proportion = data['locality_name'].isna().sum() / data['locality_name'].count()
print(f'Доля пропусков в столбце locality_name: {locality_name_proportion:.1%}')
data['locality_name'].value_counts(dropna=False)
```

```python
#так как доля пропусков мала, удалим из датафрейма строки с пропусками в названии населенных пунктов
data.dropna(subset=['locality_name'], inplace=True)
```

```python
#заменим тип данных для стоимости и дней размещения объявления на int
data['last_price'] = data['last_price'].astype('int')
data['days_exposition'] = data['days_exposition'].astype('int')
```

Пропуски оставлены в столбцах airports_nearest, parks_nearest, ponds_nearest так как поблизости отсутствуют парки, пруды и аэропорты.
В столбце cityCenters_nearest оставлены пропуски, так как они имеются в строках с домами, расположенными в маленьких населенных пунктах и разделение на центральную зону отсутствует.

<b> Вывод:</b> В ряде столбцов пропуск обозначали физическое отсутсвие данных по этому параметру, например в столбцах с количеством балконов, прудов поблизости, типом недвижимости - аппартаменты и т.п. В таком случае пропуски были оставлены или заменены на 0 в зависимости от контекста столбца. Для столбцов, пропуски в которых не имеют такую явную зависимость, пропуски были заменены на медианные или средние значения, в зависимости от диапазона и наличия выбросов. Для столбца с названием населенных пунктов, строки с пропусками были удалены, т.к доля таких строк мала = 0.2%

## Расчёты и добавление результатов в таблицу

```python
#добавление столбца с расчетом цены за квадратный метр
data['price_per_meter'] = data['last_price'] / data['total_area']
```

```python
#добавление столбцов с данными о днях недели, месяце и году размещения объявления
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d')
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
```

```python
#добавления столбца с категорией, описывающий этаж
def floor_category(row):
    if row['floor'] == row['floors_total']:
        return 'последний'
    if row['floor'] == 1:
        return 'первый'
    return 'другой'
data['floor_category'] = data[['floor', 'floors_total']].apply(floor_category, axis=1)
```

```python
#добавление столбцов с отшением кухни к общей площади и жилой площади к общей площади
data['ratio_living_total'] = ((data['living_area'] / data['total_area']) * 100).round(2)
data['ratio_kitchen_total'] = ((data['kitchen_area'] / data['total_area']) * 100).round(2)
display(data)
```
